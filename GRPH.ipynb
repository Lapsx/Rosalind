```python
def overlaping_graphs(arquivo):
    f=open(arquivo,'r')
    l=f.read().replace('\n', '').split('>')[1:]
    f.close()
    
    header = []
    dna = []
    nodes = []
    
    for i in l:
        header.append(i[:13])
        dna.append(i[13:])
        
    
    for i in range(len(dna)):
        for j in filter(lambda j:j!=i, range(len(dna))):
            if dna[i][:3]==dna[j][-3:]:
                nodes.append(header[j]+' '+header[i])

    with open('GRPH.txt', 'w') as output_data:
        output_data.write('\n'.join(nodes))
    
    return nodes
```

```python
overlaping_graphs('rosalind_grph.txt')
```
