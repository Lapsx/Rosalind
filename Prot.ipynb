```python
aminoacidos = {'UUU':'F','CUU':'L','AUU':'I','GUU':'V',
         'UUC':'F','CUC':'L','AUC':'I','GUC':'V',
         'UUA':'L','CUA':'L','AUA':'I','GUA':'V',
         'UUG':'L','CUG':'L','AUG':'M','GUG':'V',
         'UCU':'S','CCU':'P','ACU':'T','GCU':'A',
         'UCC':'S','CCC':'P','ACC':'T','GCC':'A',
         'UCA':'S','CCA':'P','ACA':'T','GCA':'A',
         'UCG':'S','CCG':'P','ACG':'T','GCG':'A',
         'UAU':'Y','CAU':'H','AAU':'N','GAU':'D',
         'UAC':'Y','CAC':'H','AAC':'N','GAC':'D',
         'UAA':'Stop','CAA':'Q','AAA':'K','GAA':'E',
         'UAG':'Stop','CAG':'Q','AAG':'K','GAG':'E',
         'UGU':'C','CGU':'R','AGU':'S','GGU':'G',
         'UGC':'C','CGC':'R','AGC':'S','GGC':'G',
         'UGA':'Stop','CGA':'R','AGA':'R','GGA':'G',
         'UGG':'W','CGG':'R','AGG':'R','GGG':'G'}
```

```python
def rna_to_prot(arquivo):
    f = open(arquivo, 'r')
    l = f.read().rstrip('\n')
    f.close()
    
    rna_m = []
    #proteina = ''
    
    for base in range(0,len(l), 3):
        #if aminoacidos[l[base:base+3]] != 'Stop':
            #proteina += aminoacidos[l[base:base+3]]
        rna_m.append(l[base:base+3])
        
    proteina = ''.join([aminoacidos.get(codon,codon)for codon in rna_m])[:-4]
    f = open('proteina.txt','w')
    f.write(proteina)
    
    return proteina
```

```python
rna_to_prot('rosalind_prot.txt')
```
